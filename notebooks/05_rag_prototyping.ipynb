```python
import sys
sys.path.insert(0, "..")
from src.retinocare.agents.retriever import HybridRetriever

retriever = HybridRetriever("../knowledge_base/guidelines")
print(f"Indexed {len(retriever.chunks)} chunks from the knowledge base.")
```

```python
test_queries = [
    "screening guidance for Mild diabetic retinopathy",
    "screening guidance for Severe diabetic retinopathy",
    "screening guidance for Proliferative DR diabetic retinopathy",
]

for query in test_queries:
    print(f"Query: {query}")
    results = retriever.retrieve(query, k=3)
    for r in results:
        print(f"  [{r['source']}] score={r['score']}")
        print(f"    {r['text'][:120]}...")
    print()
```
